```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file1 = "JC-202004-citibike-tripdata.csv"
file2 = "JC-202012-citibike-tripdata.csv"
file3 = "JC-202101-citibike-tripdata.csv"
file4 = "JC-202201-citibike-tripdata.csv"
file5 = "JC-202202-citibike-tripdata.csv"
file6 = "JC-202203-citibike-tripdata.csv"

# Read Purchasing File and store into Pandas data frame
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df3 = pd.read_csv(file3)
df4 = pd.read_csv(file4)
df5 = pd.read_csv(file5)
df6 = pd.read_csv(file6)
```

```python
dfmerge1=df1.merge(df2,how="outer")
```

```python
dfmerge2=dfmerge1.merge(df3,how="outer")
```

```python
dfmerge3=df4.merge(df5,how="outer")
```

```python
dfmerge4=dfmerge3.merge(df6,how="outer")
```

```python
dfmerge2=dfmerge2.rename(columns={"starttime":"started_at",
                        "stoptime":"ended_at",
                        "start station id":"start_station_id",
                         "start station name":"start_station_name",
                         "start station latitude":"start_lat",
                        "start station longitude":"start_lng",
                        "end station id":"end_station_id",
                         "end station name":"end_station_name",
                         "end station latitude":"end_lat",
                        "end station longitude":"end_lng",
                        "usertype":"member_casual"})
dfmerge2=dfmerge2.drop(columns=['tripduration', 'bikeid','birth year','gender',"end_station_id","start_station_id"])
```

```python

dfmerge4=dfmerge4.drop(columns=['ride_id','rideable_type',"end_station_id","start_station_id"])
```

```python
dfmerge2.head()
```

```python
dfmerge4.head()
```

```python
finalmerge=dfmerge2.merge(dfmerge4,how="outer")
```

```python
finalmerge
```

```python
stations=finalmerge.drop_duplicates(subset=['start_station_name'])[["start_station_name","start_lat","start_lng"]]
```

```python
stations
```

```python
# Import packages
import geopy

def get_zipcode(df, geolocator, lat_field, lon_field):
    location = geolocator.reverse((df[lat_field], df[lon_field]))
    return location.raw['address']['postcode']
geolocator = geopy.Nominatim(user_agent='ostrichesarecool@gmail.com')

zipcodes = stations.apply(get_zipcode, axis=1, geolocator=geolocator, lat_field='start_lat', lon_field='start_lng')
```

```python
stations["zipcodes"]=zipcodes
```

```python
finalmerge2=finalmerge.merge(stations,how="left")
```

```python
finalmerge2.to_csv('finalmerge.csv', index=False)
```
